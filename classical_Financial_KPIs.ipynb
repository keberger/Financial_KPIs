```python
df_Gas['rolling_mean_10'] = df_Gas.rolling(10).mean()
df_Gas.plot()
```

# Compute Trader Volatility of Gas prices

## Compute square deviation of gas

### Compute Log (Price(i)/ price(i-1))

```python
def compute_trader_volatility(df_prices, window):
    return pandas.DataFrame((numpy.log(df_prices/df_prices.shift(1)))/(window-1)).rolling(window).sum()*252
```

```python
df_Gas["trader_volatility"] = compute_trader_volatility(df_Gas.Price,15)
print(df_Gas.head(20))
df_Gas.plot()
```
